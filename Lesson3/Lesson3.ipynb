```python
import spacy
from spacy import displacy
import en_core_web_md
```

```python
def get_NER(df, column_name):
    nlp = en_core_web_md.load()
    lst = []
    for index, row in df.iterrows():
        doc = nlp(row[column_name])
        for ent in doc.ents:
            d = {}
            d['text'] = ent.text
            d['label'] = ent.label_
            lst.append(d)
    return lst
```

```python
l = get_NER(combine_df,'tweet')
```

```python
def getNER2(df, column_name):
    lst = []
    for index, row in df.iterrows():
        #каждый токен с большой буквы
        tokens = nltk.word_tokenize(row['tweet'].title()) 
        for chunk in nltk.ne_chunk(nltk.pos_tag(tokens)):
            if hasattr(chunk, 'label'):
                d = {}
                d['text'] = ' '.join(c[0] for c in chunk)
                d['label']= chunk.label() 
                lst.append(d)
    return lst
```

```python
l2 = getNER2(combine_df, 'tweet')
```
